####Morphology:

4. Meaning of suffix: The suffix "-ed" indicates an action that was completed in the past.

####Lexical Analysis:
1. driver: Noun (referring to a person who drives a vehicle)

####Syntax:

####Semantics:

####Pragmatics:
Implied intent if written in a 1-star review:

* If a user writes "The driver arrived on time" in a 1-star review, the implied intent is likely Fact. In a negative review, stating a positive aspect like punctuality often serves to highlight that despite this one positive, other aspects (e.g., cost, wait time) were unsatisfactory, leading to the low rating. It contrasts the punctuality with the overall negative experience, rather than being sarcastic about the driver's arrival time itself.

* So, what it actually in short means according to me is that, there was one positive thing in an overall negative 1-star experience and that was the driver "being on time".

```python
import re

input_text = data

print("Original Text:")
print(input_text)
print("\n" + "-" * 30 + "\n")

# Task 1: The Pipeline (Cleaning)
print("Task 1: Cleaning\n")

# Normalize the entire text (lowercase all letters).
normalized_text = input_text.lower()
print(f"Normalized Text (lowercase):\n{normalized_text}\n")

# Clean the text by removing special punctuation (like ... and !),
# but keep the dollar signs ($), dots for emails/prices (.),
# slashes for dates (/), at symbol for emails (@), hyphens for phone numbers (-),
# and parenthesis for phone numbers.

# First, remove '...'
cleaned_text = re.sub(r'\.{3}', '', normalized_text)

# Now, remove other unwanted punctuation, explicitly allowing for $, ., /, @, -, (, )
# The regex `[^a-z0-9 $.()/@-]+` means: match anything that is NOT a lowercase letter, digit,
# space, $, ., (, ), /, @, or -. This should preserve the necessary characters.
cleaned_text = re.sub(r'[^a-z0-9 $.()/@-]+', ' ', cleaned_text)
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip() # Replace multiple spaces with a single space

print(f"Cleaned Text (punctuation removed except $, ., /, @, -, (, )):\n{cleaned_text}\n")

print("\n" + "-" * 30 + "\n")

# Task 2: Information Extraction (Regex)
print("Task 2: Information Extraction\n")

# The Date: Find the date in DD/MM/YYYY format.
# Updated to reflect that slashes are now preserved in cleaned_text
date_pattern = r'\b(\d{2}/\d{2}/\d{4})\b'
dates = re.findall(date_pattern, cleaned_text)
print(f"Dates found: {dates}\n")

# The Price: Extract the cost (must handle the $ symbol and decimals).
price_pattern = r'\$\d+\.?\d{0,2}\b'
prices = re.findall(price_pattern, cleaned_text)
print(f"Prices found: {prices}\n")

# The Entities:
# Extract the Email Address.
# Updated to reflect that '@' and '.' are now preserved in cleaned_text
email_pattern = r'\b[a-z0-9._%+-]+@[a-z0-9.-]+\.[a-z]{2,}\b'
emails = re.findall(email_pattern, cleaned_text)
print(f"Email addresses found: {emails}\n")

# Extract the Phone Number (Note: using the Complex Pattern that handles parenthesis like (555)).
# Updated to reflect that hyphens are now preserved in cleaned_text
phone_pattern_complex = r'\((\d{3})\) (\d{3})-(\d{4})'
phone_numbers_complex = ['-'.join(m.groups()) for m in re.finditer(phone_pattern_complex, cleaned_text)]
print(f"Phone numbers (complex pattern (555) XXX-XXXX) found: {phone_numbers_complex}\n")

print("\n" + "-" * 30 + "\n")

# Task 3: Morphological Mining
print("Task 3: Morphological Mining\n")

# The Agent (The Doer): Find the word ending in -er (e.g., "driver").
agent_pattern = r'\b\w+er\b'
agents = re.findall(agent_pattern, cleaned_text)
print(f"Agents (-er words) found: {agents}\n")

# The Action (Past Tense): Find the word ending in -ed.
action_pattern = r'\b\w+ed\b'
actions = re.findall(action_pattern, cleaned_text)
print(f"Actions (-ed words) found: {actions}\n")

# The Negation: Find the word with a negative prefix (specifically un- or dis-).
# Modified to capture the full word, not just the prefix
negation_pattern = r'\b(un|dis)\w*\b'
negations = re.findall(negation_pattern, cleaned_text) # This will return only the prefix 'un' or 'dis' if no further ‘word characters’ are found.
# To get the full word, we should iterate over matches and get the whole match string.
full_negations = []
for match in re.finditer(r'\b(?:un|dis)\w+\b', cleaned_text):
    full_negations.append(match.group(0))
print(f"Negations (un- or dis- words) found: {full_negations}\n")
```
